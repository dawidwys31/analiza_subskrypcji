# Import Bibliotek

```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
# https://scikit-learn.org/stable/modules/linear_model.html#logistic-regression
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

# Wczytanie Danych

```python
df_final = pd.read_csv('../data/processed/bank_data_processed.csv', sep=',').copy()
```

# Model MLP

```python
X = df_final.drop('y', axis=1)
Y = df_final['y']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
num_cols = ['age', 'balance', 'day', 'campaign', 'pdays', 'previous']
scaler = StandardScaler()
```

```python
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
X_train_scaled[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test_scaled[num_cols] = scaler.transform(X_test[num_cols])
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(20, 10), random_state=0, max_iter=10000)
mlp.fit(X_train_scaled, Y_train)
y_pred = mlp.predict(X_test_scaled)
```

```python
metrics.confusion_matrix(Y_test, y_pred)
```

```python
print(metrics.classification_report(Y_test, y_pred))
```

# Wnioski

* Brak mechanizmu balansowania wag klas skutkuje to wysokim wskaźnikiem Accuracy, ale niskim Recall (tylko 28% skuteczności w wykrywaniu realnych subskrypcji).
* W przeciwieństwie do Regresji, sieć neuronowa jest bardziej "ostrożna" (wyższy Precision). Jeśli model przewiduje sukces, ma rację w ponad połowie przypadków.
* Model MLP działa jak "czarna skrzynka" = trudniejsze do bezpośredniego wyjaśnienia biznesowego.
